```python
import fitz  # PyMuPDF

def analyze_pdf_columns(pdf_path):
    doc = fitz.open(pdf_path)
    for page_number, page in enumerate(doc):
        text_blocks = page.get_text("dict")["blocks"]
        x_coords = [block["bbox"][0] for block in text_blocks if block["type"] == 0]  # Extract x-coordinate of each text block

        if not x_coords:  # Check if the list is empty, which means the page might be empty
            print(f"Page {page_number+1} is empty or contains no text blocks.")
            continue

        # Assuming a simple two-column layout, find median x-coordinate
        median_x = sorted(x_coords)[len(x_coords) // 2]
        
        # Classify text blocks based on their x-coordinate
        left_column = [block for block in text_blocks if block["bbox"][0] < median_x]
        right_column = [block for block in text_blocks if block["bbox"][0] >= median_x]

        print(f"Page {page_number+1}:")
        print(f"  Left Column Blocks: {len(left_column)}")
        print(f"  Right Column Blocks: {len(right_column)}")

    doc.close()
```

```python
def find_text_blocks_with_texts(lines, eps=5, min_samples=1):
    """
    Finds text block coordinates and texts from individual line coordinates and texts.
    
    :param lines: List of tuples, each containing the coordinates and the text of the line (x1, y1, x2, y2, text).
    :param eps: Maximum distance between two samples for them to be considered as in the same neighborhood.
    :param min_samples: The number of samples in a neighborhood for a point to be considered as a core point.
    :return: List of tuples, each containing the aggregated text of the block and its coordinates (block_text, x1, y1, x2, y2).
    """
    # Extracting midpoints of lines for clustering
    midpoints = [(x1 + x2) / 2 for x1, y1, x2, y2, _ in lines]

    # Clustering using DBSCAN
    clustering = DBSCAN(eps=eps, min_samples=min_samples).fit(np.array(midpoints).reshape(-1, 1))

    # Grouping lines into blocks
    blocks = {}
    for label, (x1, y1, x2, y2, text) in zip(clustering.labels_, lines):
        if label not in blocks:
            blocks[label] = {
                "texts": [],
                "coords": []
            }
        blocks[label]["texts"].append(text)
        blocks[label]["coords"].append((x1, y1, x2, y2))

    # Finding block coordinates and texts
    block_info = []
    for block in blocks.values():
        block_text = ' '.join(block["texts"])
        x1 = min(coords[0] for coords in block["coords"])
        y1 = min(coords[1] for coords in block["coords"])
        x2 = max(coords[2] for coords in block["coords"])
        y2 = max(coords[3] for coords in block["coords"])
        block_info.append((block_text, x1, y1, x2, y2))

    return block_info
```

```python
def find_text_blocks_with_texts(lines, eps=5, min_samples=1):
    """
    Finds text block coordinates and texts from individual line coordinates and texts, considering vertical white spaces.
    
    :param lines: List of tuples, each containing the coordinates and the text of the line (x1, y1, x2, y2, text).
    :param eps: Maximum distance between two samples for them to be considered as in the same neighborhood.
    :param min_samples: The number of samples in a neighborhood for a point to be considered as a core point.
    :return: List of tuples, each containing the aggregated text of the block and its coordinates (block_text, x1, y1, x2, y2).
    """
    # Extracting midpoints of lines for clustering
    points = [((x1 + x2) / 2, y1) for x1, y1, x2, y2, _ in lines]

    # Clustering using DBSCAN
    clustering = DBSCAN(eps=eps, min_samples=min_samples).fit(np.array(points))

    # Grouping lines into blocks
    blocks = {}
    for label, (x1, y1, x2, y2, text) in zip(clustering.labels_, lines):
        if label not in blocks:
            blocks[label] = {
                "texts": [],
                "coords": []
            }
        blocks[label]["texts"].append(text)
        blocks[label]["coords"].append((x1, y1, x2, y2))

    # Finding block coordinates and texts
    block_info = []
    for block in blocks.values():
        block_text = ' '.join(block["texts"])
        x1 = min(coords[0] for coords in block["coords"])
        y1 = min(coords[1] for coords in block["coords"])
        x2 = max(coords[2] for coords in block["coords"])
        y2 = max(coords[3] for coords in block["coords"])
        block_info.append((block_text, x1, y1, x2, y2))

    return block_info
```

```python
def rectangle_distance(rect1, rect2):
    """
    Calculate the distance between two rectangles.
    
    :param rect1: Coordinates of the first rectangle (x1, y1, x2, y2).
    :param rect2: Coordinates of the second rectangle (x1, y1, x2, y2).
    :return: Distance between the two rectangles.
    """
    x1_min, y1_min, x1_max, y1_max, _ = rect1
    x2_min, y2_min, x2_max, y2_max, _ = rect2

    horizontal_distance = max(x1_min - x2_max, x2_min - x1_max, 0)
    vertical_distance = max(y1_min - y2_max, y2_min - y1_max, 0)

    distance = max(horizontal_distance, vertical_distance)
    return distance
```

```python
def create_distance_matrix(rectangles):
    """
    Create a distance matrix for a list of rectangles.
    
    :param rectangles: List of rectangles (x1, y1, x2, y2, text).
    :return: A distance matrix.
    """
    n = len(rectangles)
    distance_matrix = np.zeros((n, n))

    for i in range(n):
        for j in range(n):
            if i != j:
                distance_matrix[i, j] = rectangle_distance(rectangles[i], rectangles[j])

    return distance_matrix
```

```python
def find_text_blocks(rectangles, eps=5, min_samples=2):
    """
    Cluster rectangles into text blocks, aggregate their texts, and determine block coordinates.
    
    :param rectangles: List of rectangles (x1, y1, x2, y2, text).
    :param eps: Maximum distance between two samples for them to be considered as in the same neighborhood.
    :param min_samples: The number of samples in a neighborhood for a point to be considered as a core point.
    :return: List of tuples, each containing the coordinates and aggregated text of each block (x1, y1, x2, y2, text).
    """
    distance_matrix = create_distance_matrix(rectangles)
    clustering = DBSCAN(eps=eps, min_samples=min_samples, metric="precomputed").fit(distance_matrix)

    blocks = {}
    for idx, label in enumerate(clustering.labels_):
        if label not in blocks:
            blocks[label] = {"texts": [], "coords": []}
        blocks[label]["texts"].append(rectangles[idx][4])
        blocks[label]["coords"].append(rectangles[idx][:4])

    # Aggregate texts and calculate block coordinates for each cluster
    block_info = []
    for block in blocks.values():
        aggregated_text = ' '.join(block["texts"])
        x1 = min(coord[0] for coord in block["coords"])
        y1 = min(coord[1] for coord in block["coords"])
        x2 = max(coord[2] for coord in block["coords"])
        y2 = max(coord[3] for coord in block["coords"])
        block_info.append((x1, y1, x2, y2, aggregated_text))

    return block_info
```

```python
def calculate_distances(tuples_list):
    sorted_tuples = sorted(tuples_list, key=lambda x: (x[1], x[0]))
    distances = []

    for i in range(len(sorted_tuples) - 1):
        curr = sorted_tuples[i]
        next = sorted_tuples[i + 1]

        vertical_distance = next[1] - curr[3]
        horizontal_distance = abs((next[0] + next[2])/2 - (curr[0] + curr[2])/2)

        distances.append([vertical_distance, horizontal_distance])

    return distances

def find_best_thresholds(tuples_list, num_clusters=2):
    distances = calculate_distances(tuples_list)
    distances_array = np.array(distances)

    # Apply K-Means Clustering
    kmeans = KMeans(n_clusters=num_clusters)
    kmeans.fit(distances_array)
    centroids = kmeans.cluster_centers_

    # Sort centroids (assumes the smaller distances are line spacings and larger are column gaps)
    centroids = sorted(centroids, key=lambda x: x[0])

    # Set thresholds based on centroids
    y_threshold = centroids[0][0]  # Smaller distance for line spacing
    x_threshold = centroids[1][1]  # Larger distance for column gap

    return x_threshold, y_threshold
```

```python
def is_close(cluster, line, x_threshold, y_threshold):
    # Check if the line is close to the cluster in both x and y directions
    x_close = abs((cluster[0] + cluster[2])/2 - (line[0] + line[2])/2) <= x_threshold
    y_close = abs(cluster[3] - line[1]) <= y_threshold
    return x_close and y_close

def merge_clusters(cluster, new_line):
    # Merge the new line into the existing cluster
    x1 = min(cluster[0], new_line[0])
    y1 = min(cluster[1], new_line[1])
    x2 = max(cluster[2], new_line[2])
    y2 = max(cluster[3], new_line[3])
    text = cluster[4] + " " + new_line[4]
    return (x1, y1, x2, y2, text)

def find_text_blocks(tuples_list, x_threshold=50, y_threshold=10):
    # Sort the list by y1 coordinate and then by x1 coordinate
    sorted_tuples = sorted(tuples_list, key=lambda k: (k[1], k[0]))
    clusters = []

    for tup in sorted_tuples:
        merged = False
        for i in range(len(clusters)):
            if is_close(clusters[i], tup, x_threshold, y_threshold):
                # Merge with the cluster
                clusters[i] = merge_clusters(clusters[i], tup)
                merged = True
                break
        if not merged:
            # Start a new cluster
            clusters.append(tup)

    return clusters
```

```python
import fitz  # PyMuPDF
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

def optimal_number_of_clusters(data):
    silhouette_scores = []
    for n_clusters in range(2, 10):  # Assuming a max of 10 columns, adjust as needed
        kmeans = KMeans(n_clusters=n_clusters).fit(data)
        score = silhouette_score(data, kmeans.labels_)
        silhouette_scores.append(score)
    
    optimal_clusters = silhouette_scores.index(max(silhouette_scores)) + 2  # +2 because range starts at 2
    return optimal_clusters

def get_columns(pdf_path, page_number):
    doc = fitz.open(pdf_path)
    if page_number < len(doc):
        page = doc.load_page(page_number)
        blocks = page.get_text("dict")["blocks"]
        x_coords = [[block["bbox"][0]] for block in blocks if block["type"] == 0]  # 0 is text

        if len(x_coords) > 1:
            optimal_clusters = optimal_number_of_clusters(x_coords)
            kmeans = KMeans(n_clusters=optimal_clusters).fit(x_coords)
            cluster_centers = kmeans.cluster_centers_

            print(f"Page {page_number + 1} has {optimal_clusters} columns at x-coordinates:", cluster_centers.flatten())
        else:
            print(f"Page {page_number + 1} has 1 column")
    else:
        print("Page number out of range")

    doc.close()

# Example usage
pdf_path = "a.pdf"
page_number = 10  # for the first page; adjust as needed
get_columns(pdf_path, page_number)
```
