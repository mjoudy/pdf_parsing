```python
def extract_image_from_page(page, page_number):
    image_list = page.get_images(full=True)
    images_name = []
    for img_index, img_info in enumerate(image_list):
        xref = img_info[0]
        base_image = pdf_document.extract_image(xref)
        image_bytes = base_image["image"]
        image_filename = f"image_{page_number}_{img_index + 1}.png"
        images_name.append(image_filename)
        with open(image_filename, "wb") as image_file:
            image_file.write(image_bytes)

    return images_name
```

```python
from difflib import get_close_matches

def find_section_by_title(text, target_title):
    # Convert both titles to lowercase for case-insensitive comparison
    target_title_lower = target_title.lower()
    text_lower = text.lower()

    # Split the text into lines and find the closest match to the target title
    lines = text_lower.splitlines()
    closest_match = get_close_matches(target_title_lower, lines, n=1, cutoff=0.6)

    if closest_match:
        # Get the index of the closest match in the text
        match_index = lines.index(closest_match[0])
        
        # Extract the section after the matched title
        section = '\n'.join(lines[match_index + 1:]).strip()
        return section
    else:
        return "Title not found in the text"

# Example extracted text (replace this with your actual extracted text)
example_text = """
1 
KONZEPTE, METHODEN UND GRENZEN  
DES MASCHINELLEN LERNENS
This is the content of the section.
"""

# Example target title from TOC to search for
target_title = '1\t\x07Konzepte, Methoden und Grenzen \ndes Maschinellen Lernens'  # Replace this with the title you want to find

# Find the section corresponding to the target title in the text
section = find_section_by_title(example_text, target_title)
print(section)
```

```python
import re

def split_text_by_toc(text, toc_titles):
    # Join all TOC titles into a single regex pattern
    toc_pattern = '|'.join(re.escape(title) for title in toc_titles)
    
    # Construct a regex pattern to match variations in capitalization, spaces, tabs, and new lines
    regex_pattern = r'\s*(' + toc_pattern + r')\s*'
    
    # Split the text using the constructed regex pattern
    splitted_text = re.split(regex_pattern, text, flags=re.IGNORECASE)
    
    # Remove empty strings and return the result
    return [item.strip() for item in splitted_text if item.strip()]

# Example TOC titles (update this with your actual TOC titles)
toc_titles = [
    "Introduction",
    "Chapter ONE    ",
    "Chapter Two",
    "Conclusion"
]

# Example text extracted from a PDF (replace this with your extracted text)
example_text = """
Introduction
This is the content of the introduction.
Chapter One
This is chapter one.
Chapter   Two
This is chapter two.
Conclusion
This concludes the document.
"""

# Split the text based on TOC titles
splitted_text = split_text_by_toc(example_text, toc_titles)
print(splitted_text)
```
