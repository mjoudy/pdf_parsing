```python
def extract_image_from_page(page, page_number):
    image_list = page.get_images(full=True)
    images_name = []
    for img_index, img_info in enumerate(image_list):
        xref = img_info[0]
        base_image = pdf_document.extract_image(xref)
        image_bytes = base_image["image"]
        image_filename = f"image_{page_number}_{img_index + 1}.png"
        images_name.append(image_filename)
        with open(image_filename, "wb") as image_file:
            image_file.write(image_bytes)

    return images_name
```

```python
import pprint

pp = pprint.PrettyPrinter(width=200, compact=True)
```

```python
parts_list = []

for i in range(len(rest_toc)):
    print('iii: ', i)
    title = rest_toc[i][1].casefold() 
    title = title.replace('\x07', '')
    title = title.replace('\t', '\n')
    print('tt-title: ', title) 
    start_page = rest_toc[i][2]
    print('sss-start: ', start_page)
    if i != len(rest_toc)-1:
        end_page = rest_toc[i+1][2]
    else:
        end_page = len(pdf_document)   #!probelm of out of toc pages and generalization and end page
    print('eee-end: ', end_page)
    part_text = []
    part_images = []
    for j in range(start_page-1, end_page): #page number in pdf start from 1 and in fitz from 0
        print('jjj: ', j)
        page = pdf_document.load_page(j)
        page_images = extract_image_from_page(page, j)
        if page_images is not None:
            part_images.append(page_images)
        text = page.get_text("text", sort=True).casefold()
        text = text.replace('\x07', '')
        
        #page_header, page_text = separate_header(text)
        #text = text.split("\n")[2:]   #regex is better
        #page_text = '\n'.join(text)
        part_text.append(text)   #incluse index of characers
        pp.pprint(part_text)
    
    if i != len(rest_toc) & i !=len(rest_toc)-1:   #!probelm of out of toc pages and generalization and end page
        part_text[0] = part_text[0].split(title)[1]   #to exclude text from previous part
        next_title = rest_toc[i+1][1]
        part_text[-1] = part_text[-1].split(next_title)[0]  #to exclude text from next part
    else:
        part_text[0] = part_text[0].casefold().split(title.casefold())[1]     

    part_dict = {
        'title': title,
        'pages': f'{start_page}-{end_page}',
        'text': part_text,
        'images': part_images
    }

    parts_list.append(part_dict)

#pdf_document.close()
```
